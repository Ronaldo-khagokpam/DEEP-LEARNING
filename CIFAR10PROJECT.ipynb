```python
# Importing necessary libraries
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# Step 1: Load the CIFAR-10 Dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Normalize the image data to the range [0, 1]
x_train, x_test = x_train / 255.0, x_test / 255.0

# Convert labels to one-hot encoding
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Step 2: Define the CNN Model
model = models.Sequential([
    # First convolutional layer
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),

    # Second convolutional layer
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # Third convolutional layer
    layers.Conv2D(64, (3, 3), activation='relu'),

    # Flatten the output for the fully connected layer
    layers.Flatten(),

    # Fully connected layer
    layers.Dense(64, activation='relu'),

    # Output layer with 10 neurons for 10 classes, softmax activation for classification
    layers.Dense(10, activation='softmax')
])

# Print the model summary
model.summary()

# Step 3: Compile the Model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Step 4: Train the Model
history = model.fit(x_train, y_train, epochs=10,
                    validation_data=(x_test, y_test))

# Step 5: Evaluate the Model on Test Data
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"Test accuracy: {test_acc}")

# Step 6: Plot Training and Validation Accuracy
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Test Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Step 7: Plot Training and Validation Loss
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Test Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Step 8: Make Predictions (Optional)
predictions = model.predict(x_test)
print(f"Predicted probabilities for the first test image:\n {predictions[0]}")
```

```python
# write code for image classification using CNN model.
# Load the datasets from Keras,
# Show the first 10 sample images,
# Design the model having
# 1st convolution layer - 32 filters (3x3) + relu
# Maxpool layer(2x2)
# 2nd      ,,           - 64   ,,    (3x3) + relu
# Maxpool layer(2x2)     -
# 3nd      ,,           - 64   ,,    (3x3) + relu
# For fully connected layer , 64 hidden neurons + relu
# 10 output neurons in output layer + softmax

import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow import keras

# Load the dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()

# Normalize the data
x_train = x_train / 255.0
x_test = x_test / 255.0

# Show the first 10 sample images
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(x_train[i])
    plt.title(y_train[i])
    plt.axis('off')
plt.show()

# Define the model
model = keras.Sequential([
    keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),#32 filter,3*3 kernal size,activation function relu(rectified linear unit),input shape 32*32 pixel  ,3 for RGB color
    keras.layers.MaxPooling2D((2, 2)),#maxpolling is a down szampling operation to reduce spital dimension,(2,2)-pool size specify the dimension window that moves over the features map
    keras.layers.Conv2D(64, (3, 3), activation='relu'), #2nd layer with64 filter and kernal size 3*3 with RELU activation
    keras.layers.MaxPooling2D((2, 2)),#2nd maxpolling
    keras.layers.Conv2D(64, (3, 3), activation='relu'),#3rd convolution layer
    keras.layers.Flatten(),#to reshape the data of one dimension for fully connected layer
    keras.layers.Dense(64, activation='relu'),#A dense layer is a layer where each neuron is connected to every neuron in the previous layer
    keras.layers.Dense(10, activation='softmax')

    """keras.layers.Dense:This creates a fully connected layer, also known as a dense layer, where each neuron in the layer is connected to every neuron in the previous layer.

64:This parameter specifies the number of neurons in the dense layer. In this case, the layer has 64 neurons.

activation='relu':The activation parameter defines the activation function to be used by the neurons in this layer. 'relu' stands for Rectified Linear Unit, which is a common activation function used in neural networks to introduce non-linearity."""
])

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(x_train, y_train, epochs=10)

# Evaluate the model
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print('\nTest accuracy:', test_acc)
```
